```python
import pandas as pd
import numpy as  np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('intern.csv',delimiter=',',skiprows=4)
df
```

```python
countries=['Aruba','Algeria','Afghanistan','Cuba','Finland','Fiji','Greece','India']
year ='2023'
filtered_data = df[df['Country Name'].isin(countries)][['Country Name', year]]

plt.figure(figsize=(10, 6))
plt.bar(filtered_data['Country Name'], filtered_data[year], color='Red',edgecolor='black')
plt.title('Distribution of some country in the year 2022',fontweight='bold')
plt.xlabel('COUNTRY NAME',fontweight='bold')
plt.ylabel('YEAR',fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(filtered_data[year], bins=5,color='blue',edgecolor='black')
plt.title('Histogram of Population Distribution of some countries in 2023',fontweight='bold')
plt.xlabel('Populatin in 2023',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(filtered_data[year], bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Population in 2020')
plt.ylabel('Frequency')
plt.title('Histogram of Population Distribution in 2020 for Selected Countries')
plt.tight_layout()

# Show the plot
plt.show()
```
